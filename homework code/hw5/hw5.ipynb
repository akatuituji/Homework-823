```python
import requests
import pandas as pd
import numpy as np
#Fetch
resp = requests.get('https://swapi.dev/api/people')
data = resp.json()
df = pd.json_normalize(data)
df
```

```python
df1= pd.DataFrame()
i = 1
while len(df1) < 82:    
    resp = requests.get('https://swapi.dev/api/people/' + str(i))
    status_code = resp.status_code
    if status_code == 200:
        data = resp.json()
        df2 = pd.json_normalize(data)
        df1 = df1.append(df2)
    i += 1
```

```python
df1
```

```python
df1 = df1.set_index(['name'])
```

```python
df1["birth_year"] = df1["birth_year"].replace('unknown',np.nan)
birth_year = pd.DataFrame(df1.birth_year)
birth_year['age'] = (birth_year.birth_year.str.replace('BBY','').astype('float'))
birth_year = birth_year.sort_values(['age'],ascending = False)
```
