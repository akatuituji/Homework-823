### Second Normal Form (2NF) and Third Normal Form (3NF)

```python
df_track = pd.concat([df.iloc[:,0:4],df.iloc[:,[4,8]],df.iloc[:,11:]],axis=1).drop_duplicates()
df_track.reset_index(drop=True, inplace=True)

df_track
```

```python
df_album_name = df.iloc[:,[4,5,6]].drop_duplicates()
df_album_name.reset_index(drop=True, inplace=True)
df_album_name.set_index('track_album_id', append=True)
df_album_name.reset_index(level=0, inplace=True)
df_album_name = df_album_name.iloc[:,[0,1]]
df_album_name
```

```python
left=df_track
right=df_album_name
df1=pd.merge(left, right, on='track_album_id',how='inner')
df1=df1.rename(columns={'index':'album'})
df1=df1.drop('track_album_id',axis=1)
df1
```

```python
df_playlist = df.iloc[:,7:11].drop_duplicates()
df_playlist.reset_index(drop=True, inplace=True)
df_playlist.set_index('playlist_id', append=True)
df_playlist.reset_index(level=0, inplace=True)
df_playlist
```

```python
df_playlist1=df_playlist
df_playlist1 = df_playlist1.iloc[:,[0,2]]
df_playlist1
```

```python
left=df1
right=df_playlist1
df2=pd.merge(left, right, on='playlist_id',how='inner')

df2=df2.rename(columns={'index':'playlist'})
df2=df2.drop('playlist_id',axis=1)
df2
```

### find the names of all playlists that contain instrumentals

```python
df_playlist=df_playlist.rename(columns={'index':'playlist_num'})
df2.to_sql('spotify', conn, if_exists='append', index=False) 
df_playlist.to_sql('df_playlist', conn, if_exists='replace', index = False)
```

```python
c = conn.cursor()
c.execute("SELECT * FROM sqlite_master WHERE type='table'")
c.fetchall()
```

```python
query = '''
SELECT df_playlist.playlist_name, track_name,instrumentalness
FROM spotify
INNER JOIN
df_playlist
ON spotify.playlist=df_playlist.playlist_num
WHERE spotify.instrumentalness > 0 -- a query contains instrumentals when its instrumentalness is greater than zero
ORDER BY spotify.instrumentalness DESC
'''
```

```python
df3=pd.read_sql(query, conn)
df3=df3.drop_duplicates()
df3
```
